```python
import pandas as pd
import random
import datetime

# Sample data
drivers = ["John Doe", "Alice Smith", "Michael Johnson", "Sarah Brown", "David Wilson"]
staff_names = ["Emma", "James", "Olivia", "Liam", "Sophia", "Noah", "Ava", "Ethan", "Isabella", "Mason"]
locations = ["Lagos", "Abuja", "Port Harcourt", "Kano", "Enugu", "Ibadan", "Kaduna"]
companies = ["ZoomCar", "BoltDrive", "GreenWheels", "QuickRide", "GoCar"]

# Generate 50 rows of sample data
rows = []
for _ in range(500):
    # Random date
    year = 2025
    month = random.randint(1, 12)
    day = random.randint(1, 28)  # keep it safe for February
    
    driver = random.choice(drivers)
    staff = random.choice(staff_names)
    from_loc, to_loc = random.sample(locations, 2)
    
    odo_start = random.randint(10000, 50000)
    odo_finish = odo_start + random.randint(10, 300)
    
    start_hour = random.randint(6, 12)
    start_minute = random.choice([0, 15, 30, 45])
    start_time = datetime.time(start_hour, start_minute)
    
    duration_hours = random.randint(1, 6)
    finish_time = (datetime.datetime.combine(datetime.date.today(), start_time) +
                   datetime.timedelta(hours=duration_hours)).time()
    
    cost = round(random.uniform(20, 200), 2)
    company = random.choice(companies)
    
    rows.append([day, month, year, driver, staff, from_loc, to_loc, odo_start, odo_finish,
                 start_time.strftime("%H:%M"), finish_time.strftime("%H:%M"), cost, company])

# Create DataFrame
columns = ["Day", "Month", "Year", "Driver", "Staff Name", "From", "To", "Odometer Start",
           "Odometer Finish", "Start Time", "Finish Time", "Cost", "Car hire Company"]
df = pd.DataFrame(rows, columns=columns)

df.to_excel('customer_hire.xlsx', index=False)
```

```python
import pandas as pd
import random
import numpy as np

# Define possible values for categorical fields
approved_status = ["Approved", "Not Approved"]
travel_types = ["Local", "International"]
cleared_status = ["Cleared", "Not Cleared"]
states = ["Lagos", "Abuja", "Port Harcourt", "Kano", "Enugu"]
feedback_options = ["Excellent", "Good", "Average", "Poor"]

# Generate sample data
num_rows = 20
data = {
    "Date of Travel Request": pd.date_range("2025-08-01", periods=num_rows, freq="D"),
    "Date of Booking Confirmation": pd.date_range("2025-08-02", periods=num_rows, freq="D"),
    "Approved or not": [random.choice(approved_status) for _ in range(num_rows)],
    "Actual Budget": [random.randint(500, 5000) for _ in range(num_rows)],
    "Approved Budget": [random.randint(500, 5000) for _ in range(num_rows)],
    "Expected travel date": pd.date_range("2025-08-05", periods=num_rows, freq="D"),
    "Actual travel date": pd.date_range("2025-08-06", periods=num_rows, freq="D"),
    "Vendor Service Request Date": pd.date_range("2025-08-03", periods=num_rows, freq="D"),
    "Vendor Service Confirmation Date": pd.date_range("2025-08-04", periods=num_rows, freq="D"),
    "Traveler Feedback": [random.choice(feedback_options) for _ in range(num_rows)],
    "Travel Type": [random.choice(travel_types) for _ in range(num_rows)],
    "Travel Cleared or not": [random.choice(cleared_status) for _ in range(num_rows)],
    "State": [random.choice(states) for _ in range(num_rows)],
}

df = pd.DataFrame(data)

df.to_excel('travel-approval.xlsx', index=False)
```
